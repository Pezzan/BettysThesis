```python
def FindThreshold(FFT, alpha):
    TH = np.mean(FFT)
    ref = 0.1
    totAmp = np.sum(FFT)
    diff = 100
    
    while (np.abs(diff) > ref):
        _, thresh = cv2.threshold(FFT, TH, 1,
                                  cv2.THRESH_BINARY)
        relAmp = np.sum(FFT*thresh)
        # number of bins over the TH
        n = np.sum(thresh)
    
        diff = (alpha*totAmp - relAmp) / n
        
        TH += (-1 * diff /50)
        print(TH, diff, totAmp, relAmp, n)
        
    return TH
    
```

```python
img = cv2.imread("For_ANALYSIS/135TCB_CB/135TCB_CB_sample1.jpg")

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
dftIm = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT)
dftImShift = np.fft.fftshift(dftIm)
magIm = 20*np.log(cv2.magnitude(dftImShift[:,:,0],dftImShift[:,:,1]))

TH = FindThreshold(magIm, 1/50)

_, thresh = cv2.threshold(magIm, TH, 1,
                            cv2.THRESH_BINARY)

plt.imshow(thresh)
```
