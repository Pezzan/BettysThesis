```python
import dill
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.image as mpimg
from PIL import Image
import cv2
```

```python
folder = 'results_discussion/135TCB_CB/'

#n = [0, 6, 7, 8, 1, 2, 3, 9, 10, 11, 4, 5]
n = [0, 4, 5, 6, 10, 11, 1, 2, 3, 7, 8, 9]
largeIdx = [0, 1, 2, 6, 7, 8]
files = glob.glob(folder + "*.jpg")
files.sort()

a = 2.7
gs = gridspec.GridSpec(4, 3,width_ratios=[a,a,a], height_ratios=[a, 1, a, 1])

f = plt.figure(figsize = (6.3,12))
for i in range(len(files)):
    plt.subplot(gs[i])
    if (i in largeIdx):
        img = mpimg.imread(files[n[i]])
        temp = mpimg.imread(files[0])
        size = temp.shape
        img = cv2.resize(img, dsize=(size[1],size[0]), interpolation=cv2.INTER_CUBIC)
    else:
        img = mpimg.imread(files[n[i]])
        #temp = mpimg.imread(files[12])
        #size = temp.shape
        #img = cv2.resize(img, dsize=(size[1],size[0]), interpolation=cv2.INTER_CUBIC)
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    
plt.subplots_adjust(wspace=0, hspace=0)
plt.savefig('figures/cameraFibers.jpeg', dpi=1000)
plt.show()
```
